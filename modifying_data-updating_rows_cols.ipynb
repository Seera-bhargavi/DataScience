```python
import pandas as pd
```

```python
people={
    "first":["janu","bharu","rihana","sai"],
    "last":["seera","seera","shaik","kola"],
    "email":["janu@gmail.com","bharu@gmail.com","rihana@gmail.com","sai@gmail.com"]
}
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
# get the columns
df.columns
```

```python
# renaming the columns
df.columns=['First_name','Last_name','email']
```

```python
df
```

```python
#changing columns names to upper-case
df.columns=[x.upper() for x in df.columns]
```

```python
df
```

```python
#s string manipulation i.e, replacing underscore(_) with @ symbol
df.columns=df.columns.str.replace('_','@')
df
```

```python
# converting lables into original labels (rearranging the labels)

df.columns=df.columns.str.replace('@',' ')
df.columns=[x.lower() for x in df.columns]
df
```

```python
# rename the specific columns
# inplace = True (keep the changes in same variable)
df.rename(columns={'first name':'first','last name':'last'}, inplace=True)
```

```python
df
```

```python
# updaing the row using loc
# syntax : DataFrame.loc[index]=['value','value',]
# make sure that columns and values should be equal
df.loc[1]=['Bhargavi','seera','seera@gmail.com']
df
```

```python
# Let's update the value of row last and email
# syntax: DataFrame.loc[index,[column_names]]

df.loc[3,['last','email']]=['kolla','saikolla@gmail.com']
df
```

```python
# Now let's update cell value with index 0 and Column first
# We will replace value janu with jhansi

df.loc[0,['first']]='jhansi'
df
```

```python
df.loc[0,['first']]
```

```python
# You can access cell value via .loc but you can't updated it this way!
# df.loc[index].at['column']

df.loc[0].at['first']='janu'
df
```

```python
# Using filter we can update the value,here we are updating the first row of last column value

filt=(df['email']=='seera@gmail.com')
df.loc[filt,'last']='SEERA'
df
```

```python
df.loc[1].at['email']='SEERA@gmail.com'
df
```

```python
#changing the email column to lowercase

df['email']=df['email'].str.lower()
df
```

```python
# To capitalize the first letter, use the capitalize() function.This functions converts the first character to uppercase and converts the remaining characters to lowercase.

df['first']=df['first'].str.capitalize()
df
```

### apply():

```python
# apply(): This method is used to call function on values
#eg: we'll see apply() method on series i.e applying length function on email column, returns the length of each emailId

df['email'].apply(len)
```

```python
# we can also use userdefined functions on apply()

def upper_fun(email):
    return email.upper()
```

```python
df['email']=df['email'].apply(upper_fun)
df
```

```python
# Now we'll change these uppercase emails to lowercase using lamda function

df['email']=df['email'].apply(lambda x: x.lower())
df
```

### now we'll see apply() method on dataframe

```python
# Applying len function on dataframe returns entries in each column
# first has 4 rows,last has 4 rows and so on..

df.apply(len)
```

```python
# Applying min() on DataFrame, checks the ascii value of each first character returns minimum ascii valued character

df.apply(lambda x:x.min())
```

```python
#applymap(): Applies function in each value in a dataframe
df.applymap(len)
```

```python
df['first']=df['first'].replace({'janu':'jhansi','sai':'surya','bhargavi':'bharu'})
```

```python
df['last']=['seera','seera','shaik','kolla']
df
```

```python
# Concatenating the 'first' column and 'last' column

df['full_name']=df['first']+' '+df['last']
df
```

```python
df.loc[0].at['last']='xxxx'
df.loc[0].at['full_name']=df.loc[0,'first']+df.loc[0,'last']
df
```

#### Removing columns using drop()

```python
#to remove a columns -- DataFrame.drop(columns=[column_names])

df.drop(columns=['first','last'])
```

```python
# Splitting the full_name using split()

df['full_name']=df.loc[[1,2,3],['full_name']].split(' ',expand=True)
```

#### Adding rows

```python
#creating dataframe
new_row = {'first':['haritha','ganga'],'last':['gumm','err'],'email':['haritha@gmail.com','ganga@gmail.com']}

r1=pd.DataFrame(new_row)
r1
```

```python
#creating another dataframe
new_row = {'first':['kalpana','kanchana'],'last':['kill','horror'],'email':['kalpana@gmail.com','kanchana@gmail.com']}
r2=pd.DataFrame(new_row)
r2
```

#### Concatnating two dataframes

```python
# Using Concat() funtion
# Syntax : pandas.concat([DataFrame1,DataFrame2],ignore_index=True)

df=pd.concat([r1,r2],ignore_index=True)
```

```python
df
```

#### Removing Rows using filter

```python
# Selecting column having value err and removing that row 

filt=df['last']=='err'

df.drop(index=df[filt].index)
```
